# **Lógica quiz**

```python
class QuizSenhorDosAneis:
    def __init__(self):
        self.pontuacao = 0
        self.nivel = 1
        self.sair = False
        self.contador_pergunta = 0

    def apresentar_instrucoes(self):
        print('Bem-vindo ao Quiz do Senhor dos Anéis!')
        print('Acerte as questões para subir de nível.')
        print('Quando chegar a um certo nível, as questões ficarão mais difíceis.')
        print('Se quiser parar, digite "S" a qualquer momento para sair.\n\n\n')

    def realizar_pergunta(self):
        if self.sair:
            self.informar_finalizacao()
            return

        if self.nivel == 1:
            self.percorre_perguntas(questoes_basicas)
        elif self.nivel == 2:
            self.percorre_perguntas(questoes_intermediarias)
        elif self.nivel == 3:
            self.percorre_perguntas(questoes_avancadas)
        else:
            self.informar_vitoria()
            self.sair = True
            return

    def percorre_perguntas(self, perguntas):
        for alternativa in perguntas:
            self.contador_pergunta += 1
            print(f'{self.contador_pergunta}) {alternativa["pergunta"]}\n')
            for i in range(len(alternativa['opcoes'])):
                print(f'{letras_opcoes[i]}) {alternativa["opcoes"][i]}')
            resposta_usuario = input("\nSua resposta (ou 'S' para sair): ").upper()

            if resposta_usuario == "S":
                print("Você decidiu sair do Quiz. Até logo!")
                self.sair = True
                return

            if resposta_usuario == self.retorna_questao_correta(alternativa['opcoes'], alternativa['resposta_correta']):
                print("Resposta correta! Você ganhou 10 pontos.\n\n")
                self.pontuacao += 10
                self.verificar_nivel()
            else:
                print(f"Resposta incorreta. A resposta correta era {alternativa['resposta_correta']}.")
                self.informar_finalizacao()
                self.sair = True
                return

    def retorna_questao_correta(self, opcoes, opcao_correta):
        indice = opcoes.index(opcao_correta) if opcao_correta in opcoes else -1

        if indice == -1:
            return 'X'
        else:
            return letras_opcoes[indice]

    def informar_finalizacao(self):
        print(f'Obrigado por jogar, sua pontuação foi de {self.pontuacao} pontos. Você parou no nível {self.nivel}.')

    def informar_vitoria(self):
        print(f'Parabéns! Você acertou todas as respostas! Obrigado por jogar, sua pontuação foi de {self.pontuacao} pontos. Você parou no nível {self.nivel}.')

    def verificar_nivel(self):
        if self.pontuacao >= self.nivel * 70:
            self.nivel += 1
            if self.nivel <= 3:
                print(f"Parabéns! Você subiu para o Nível {self.nivel}.\n")

    def iniciar_simulacao(self):
        self.apresentar_instrucoes()

        while not self.sair:  # Limite de níveis para este exemplo
            self.realizar_pergunta()
```
